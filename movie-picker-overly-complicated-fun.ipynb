```python
movie_url = "https://docs.google.com/spreadsheets/d/16vRRprQ6Fe0zFZT4rpdEocVF2QNEvPFyoufBZ9oRVwU/edit?usp=share_link"
movie_data = requests.get(movie_url).text
movie_soup = BeautifulSoup(movie_data)
movie_tbody = movie_soup.find_all('tbody')

movie_df = pd.DataFrame(columns=["movies_to_watch", "proposed_additions", "watched"])

for row in movie_soup.tbody.find_all("tr"):
    col = row.find_all("td")
    if (col != []):
        movies_to_watch = col[0].text
        proposed_additions = col[1].text
        watched = col[3].text
        movie_df = movie_df.append({"movies_to_watch":movies_to_watch, "proposed_additions":proposed_additions, "watched":watched}, ignore_index=True)

del movie_df["proposed_additions"]
del movie_df["watched"]
```
